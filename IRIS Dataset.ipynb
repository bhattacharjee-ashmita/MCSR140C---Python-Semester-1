```python
# Histogram of all features/columns
# Histogram is a univariate plot
plt.figure(figsize=(10,15)) 
plt.tight_layout()
df.hist(bins=10)
```

```python
# Convert Species column to numeric. 'Iris-Setosa' =0, 'Iris-Versicolor'=1, 'Iris-Virginica'=2
d = {'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}

df['target'] = df['Species'].map(d)      # Create a column 'target' 0 for Setosa, 
                                         # 1 for Versicolor, 2 for Virginica

print(df.head())
print(df.Species.unique())
print(df.target.unique())
```

```python
colors = df.target
plt.scatter(df['SepalLength'], df['SepalWidth'], c=colors)
plt.title('Sepal Length vs SepalWidth')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```

```python
#df.plot(kind ="scatter", x ='PetalLength', y ='PetalWidth')
colors = df.target
plt.scatter(df['PetalLength'], df['PetalWidth'], c=colors)
plt.title('Petal Length vs Petal Width')
 
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.grid()
```

```python
# Boxplot using Seaborn
plt.figure(figsize=(15,10)) 
plt.subplot(2,2,1)
sns.boxplot( x=df["Species"], y=df["SepalLength"] )    
plt.title('Boxplot Sepal Length vs Species')
plt.subplot(2,2,2)
sns.boxplot( x=df["Species"], y=df["SepalWidth"] )    
plt.title('Boxplot Sepal Width vs Species')
plt.subplot(2,2,3)
sns.boxplot( x=df["Species"], y=df["PetalLength"] )   
plt.title('Boxplot Petal Length vs Species')
plt.subplot(2,2,4)
sns.boxplot( x=df["Species"], y=df["PetalWidth"] )    
plt.title('Boxplot Petal Width vs Species')
plt.show()
```

```python
# Countplot 
sns.countplot(x=df.Species, data=df)
```

```python
#Scatter Plot with legend using Seaborn
sns.FacetGrid(df, hue ="Species", height = 4).map(plt.scatter, 'SepalLength','SepalWidth').add_legend()
```
