```python
# Read the CSV file into DF called empdf1
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
empdf1 = pd.read_csv('empmst.csv', index_col ='Empno', parse_dates=['DOB'])
empdf1
```

```python
print(empdf1.iloc[[1,2]])		# Select row 1,2
print(empdf1.iloc[[0,3]])		# Select row 0,3
# Use slice operator to select a range
print(empdf1.iloc[0:3])		# Select row 0,1,2 but not 3. Please note this is one pair of [ ]
	
```

```python
print(empdf1.iloc[0:3, [0,1,2]])		# Select rows 0,1,2 and columns 0,1,2. 
```

```python
print(empdf1.iloc[[1,4], [0,2]])		# Select row 1,4 with columns 0,2
print(empdf1.iloc[:,1])			# Select column 1 (i.e. DOB) for all rows
print(empdf1.iloc[:, -1])			# Last column (HRA) of all rows

```

```python
print(empdf1.loc[:, 'Empname'])                               # Print Empname for all rows because ':' in row position
print(empdf1.loc[67120, 'Empname'])                           # Selects Empname for Empno=67120 
print(empdf1.loc[67120, ['Empname', 'Salary']])               # Print Empname and Salary of  Empno=67120
print(empdf1.loc[[20101, 67120], ['Empname', 'Salary']])      # Empname & Salary for Empno=20101, 67120
```

```python
empdf1.info()
```

```python
empdf1.tail()
```

```python
empdf1['Salary'] = empdf1['Salary'].replace({'#':''}, regex=True) 
empdf1['Salary'] = empdf1['Salary'].replace({'@':''}, regex=True) 
empdf1['HRA'] = empdf1['HRA'].replace({"'":''}, regex=True) 
empdf1.tail()
```

```python
empdf1.info()
```

```python
empdf1 = empdf1.dropna()
empdf1.info()
```

```python
# Converting Salary and HRA from Object to float
empdf1['Salary'] = empdf1['Salary'].astype(float)
empdf1['HRA'] = empdf1['HRA'].astype(float)
empdf1.info()
```

```python
# Now Add a column Total = Salary + HRA
empdf1['Total'] = empdf1['Salary'] + empdf1['HRA']
empdf1.tail()
```

```python
# Selecting rows of a DF by Boolean conditions
print(empdf1.Salary > 50000)         # Prints Empno with 'True'/ 'False' depending on the condition.
print(empdf1[empdf1.Salary > 50000])  # Prints rows with all columns where condition is True
# Multiple condition 
print('Employees with Salary < 75000 and >= 50000')
print(empdf1[(empdf1.Salary < 75000) & (empdf1.Salary >=50000)])
```

```python
# Renaming columns 
empdf1=empdf1.rename(columns={'Salary':'EmpSalary'})   # Salary changed to EmpSalary
empdf1.tail()   
```

```python
# Make another copy of the DF
empcopy = empdf1.copy()      # Made a copy of the original DF
empcopy.tail()
```

```python
# Add column called 'Adhoc' as 30% of EmpSalary at end
empdf1.head()
# Insert a column at a position using 'insert()'. 
empdf1.insert(7, 'DA', empdf1.EmpSalary * 0.10)      # Col DA (10% of Salary) added at position 7 (after EmpSalary)
empdf1.head()
```

```python
# Delete Columns
empdf1.head()
del empdf1['DA']        # Delete the column DA
del empdf1['Adhoc']
empdf1.head()           # Prints the DF without DA.  
```

```python
empdf1['City'].replace('Bangalore', 'Bengaluru', inplace=True)
empdf1['City'].replace('Mangalore', 'Mangaluru', inplace=True)
empdf1 
```

```python
# Adds a row in the DF with Empno(90990), Empname(Mainak Gupta), DOB(16-11-1991), City(Kolkata),State(West Bengal) 
# ExpYr (5.5 yrs), EmpSalary(45000), HRA(10000), 0
empdf1.loc[90990] = ['Mainak Gupta', '16-11-1998', 'M', 'Chennai', 'Tamil Nadu', 5.5, 45000, 4500, 10000, 0]    # Changes original DF
empdf1.tail()

 
```

```python
# Drop a record based on index Empno
empdf1.drop([90990], inplace=True)
empdf1.tail()
```

```python
# Now save this DF onto a disk file called 'empmst_new.csv' in .csv format
empdf1.to_csv('empmst_new.csv', index_label='Empno')   # This new file would now contain HRA, Total also
```

```python
citywise_df = empdf1.groupby(['City'])['EmpSalary','HRA','Total'].sum()
print('City Wise Figures')
print(citywise_df)
```

```python
# Draw a bar chart 
import matplotlib.pyplot as plt
citywise_df.plot(kind='bar')       #, color = ['coral', 'yellow', 'black'])
plt.xlabel('City-->')
plt.ylabel('Remuneration -->')
plt.title('City Wise Remuneration', color='gold', fontname='Arial', fontsize=20)
plt.show(),  
```

```python
# # Find out state wise, city wise sum of salary, hra and total
# And plot a Bar chart state wise
statecitywise_df = empdf1.groupby(['State','City'])['EmpSalary','HRA','Total'].sum()
print('State Wise City Wise Total Figures')
print(statecitywise_df)
state_df = empdf1.groupby(['State'])['EmpSalary', 'HRA', 'Total'].sum()
state_df.plot(kind='bar')
plt.xlabel('State -->')
plt.ylabel('Remuneration')
plt.title('State Wise Figures')
plt.show()
```

```python
# Calculation of mean (Average) figures
 
statewise_meandf = empdf1.groupby(['State'])['EmpSalary', 'HRA', 'Total'].mean()
print('State Wise Mean figures')
print(statewise_meandf)
statewise_meandf.plot(kind='bar', figsize=(6,6))       # figsize to control the size of the chart
plt.xlabel('State -->')
plt.ylabel('Remenuration')
plt.title('State Wise Mean Figures')
plt.legend(loc='upper center', ncol=3, title="Remuneration Mean Figures")
plt.show()
```

```python
# Gender wise figures in a stacked bar chart
gender_df = empdf1.groupby(['Sex'])['HRA', 'EmpSalary', 'Total'].sum()
print('Gender Wise Figures')
print(gender_df)

gender_df.plot(kind='bar', figsize=(4,4), stacked=True)
plt.xlabel('Gender -->')
plt.ylabel('Remuneration-->')
plt.title('Gender Wise Figures')

#plt.show()
```

```python
# Scatter Plot showing Linear Regression between Salary and ExpYr
import matplotlib.pyplot as plt
expyr = empdf1['ExpYr']       
salary = empdf1['EmpSalary']    
plt.scatter(expyr, salary, color = 'red', label='Salary')
plt.title('Salary vs Experience\nScatter Plot', color='black', fontsize=30, fontname='Arial')
plt.xlabel('Years of Experience-->')
plt.ylabel('Salary-->')
plt.legend()
```
